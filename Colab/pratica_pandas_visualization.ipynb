Através dessa criação você pode explorar diversas possibilidades e reconhecer de forma atrativa o seu processo de desenvolvimento.E não nos esqueçamos das orientações! Essa figura precisa ter uma linha para cada país, título, rótulos nos eixos, cores apropriadas, um tema da biblioteca Seaborn e legenda. Por isso, pense nas questões de acessibilidade, como tamanho das fontes e espessura das linhas. É importante escolher cores adequadas que não causem cansaço visual ou dificultem a leitura das informações.
```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator

# Lista de anos
anos = list(map(str, range(1980, 2014)))

# Usa o DataFrame correto
df = df_paises

# Melt
df_long = df.melt(
    id_vars=['País'],
    value_vars=anos,
    var_name='Year',
    value_name='Value'
)

# Selecionar países fixos
paises_fixos = ['Brasil', 'Argentina', 'Peru', 'Colômbia']

df_top4 = df_long[df_long['País'].isin(paises_fixos)]

# Criar figura única
plt.figure(figsize=(14, 6))
sns.set_style("whitegrid")

cores = sns.color_palette("Set2", 4)

for pais, cor in zip(paises_fixos, cores):
    dados = df_top4[df_top4['País'] == pais]
    anos_int = dados['Year'].astype(int)

    plt.plot(anos_int, dados['Value'], label=pais, linewidth=2.5, color=cor)

plt.title("Imigração para o Canadá (1980–2013)\nBrasil, Argentina, Peru e Colômbia", fontsize=18, loc='left')
plt.xlabel("Ano")
plt.ylabel("Número de Imigrantes")

# Mostrar anos de 5 em 5
plt.gca().xaxis.set_major_locator(MultipleLocator(5))

plt.xticks(rotation=45)
plt.legend(title="País")
plt.tight_layout()
plt.show()
```

Lembra que nós criamos uma figura estática contendo os dados de imigração do Brasil e Argentina? Sua tarefa é criar um gráfico animado com o Plotly que mostre esses dados. O gráfico deve ter as seguintes características:

Duas linhas: uma para o Brasil e outra para a Argentina.
Um botão "Play" para iniciar a animação, mostrando o aumento ou diminuição do número de imigrantes ao longo dos anos.
As configurações de animação devem fazer com que as duas linhas sejam exibidas e animadas ao mesmo tempo.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, HTMLWriter

# --- PREPARAR OS DADOS ---
df = df_paises

anos = list(map(str, range(1980, 2014)))

df_long = df.melt(
    id_vars=['País'],
    value_vars=anos,
    var_name='Ano',
    value_name='Imigrantes'
)

# filtrar Brasil e Argentina
df_anim = df_long[df_long['País'].isin(['Brasil', 'Argentina'])]
df_anim['Ano'] = df_anim['Ano'].astype(int)

# --- CONFIGURAR FIGURA ---
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xlim(1980, 2013)
ax.set_ylim(0, df_anim['Imigrantes'].max() * 1.1)
ax.set_title("Imigração Brasil x Argentina para o Canadá (1980–2013)")
ax.set_xlabel("Ano")
ax.set_ylabel("Imigrantes")

# linhas vazias (serão atualizadas na animação)
linha_brasil, = ax.plot([], [], lw=3, color="green", label="Brasil")
linha_argentina, = ax.plot([], [], lw=3, color="blue", label="Argentina")

ax.legend()

# --- FUNÇÃO DE ANIMAÇÃO ---
def animate(frame):
    anos_plot = list(range(1980, 1980 + frame))

    # Brasil
    dados_brasil = df_anim[df_anim["País"] == "Brasil"]
    valores_brasil = dados_brasil[dados_brasil["Ano"].isin(anos_plot)]["Imigrantes"]
    linha_brasil.set_data(anos_plot, valores_brasil)

    # Argentina
    dados_arg = df_anim[df_anim["País"] == "Argentina"]
    valores_arg = dados_arg[dados_arg["Ano"].isin(anos_plot)]["Imigrantes"]
    linha_argentina.set_data(anos_plot, valores_arg)

    return linha_brasil, linha_argentina

# número de frames = quantidade de anos
frames = len(anos)

anim = FuncAnimation(fig, animate, frames=frames, interval=150, blit=True)

# --- SALVAR COMO HTML ---
writer = HTMLWriter(fps=5)
anim.save("animacao_brasil_argentina.html", writer=writer)

plt.close()  # evita exibir no Jupyter
```
