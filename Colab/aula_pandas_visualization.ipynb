```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import MultipleLocator
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
```

```python
df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
# analizar tendencias de imigraçao
df.set_index('País', inplace=True)
df.head()
```

```python
# armazenar o tempo numa variavel
anos = list(map(str, range(1980,2014)))
anos
```

```python
# pegar somente dados do Brasil
brasil = df.loc['Brasil', anos]
brasil
```

```python
# novo nome para colunas
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
# grafico de linhas
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigrações do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrações')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
# criação de figuras
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigrações do Brasil para o Canadá\n 1980-2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrações')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
# dois gráficos juntos
fig, axs = plt.subplots(1,2, figsize=(15,5))
# primeiro gráfico
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigrações do Brasil para o Canadá\n 1980-2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrações')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
# segundo gráfico
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot Imigrações do Brasil para o Canadá\n 1980-2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrações')
axs[0].grid()
```

```python
# vários gráficos na mesma figura com subplot
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigrações dos 4 maiores países da América do Sul')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrações')




ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette= palette)
  ax.set_title('Países com maior imigração para o Canadá\n 1980-2013', loc='left', fontsize = 18)
  ax.set_xlabel('Número de Imigrações', fontsize = 14)
  ax.set_ylabel('')
  plt.show()
```

```python
df_america_sul_clean = america_sul.drop(['Continente', 'Região', 'Total'], axis=1)
america_sul_final = df_america_sul_clean.T
america_sul_final
```

```python
import plotly.express as px

# converter índice para coluna "Ano"
df_reset = america_sul_final.reset_index().rename(columns={'index': 'Ano'})

# Melt: transforma colunas de países em linhas
df_long = df_reset.melt(
    id_vars='Ano',
    var_name='País',
    value_name='Imigrantes'
)

# Gráfico
fig = px.line(
    df_long,
    x='Ano',
    y='Imigrantes',
    color='País',
    title='Imigração dos países da América do Sul para o Canadá (1980–2013)',
    markers= True
)

fig.update_layout(
    xaxis={'tickangle': -45},
    xaxis_title='Ano',
    yaxis_title='Número de imigrantes'
)

fig.show()
```
