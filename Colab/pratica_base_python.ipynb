6) Escreva um programa que leia três números e os exiba em ordem decrescente.

```python
try:
    # 1. CONVERSÃO: Converte as entradas de string para número inteiro (int)
    num1 = int(input('Digite o primeiro número: '))
    num2 = int(input('Digite o segundo número: '))
    num3 = int(input('Digite o terceiro número: '))

except ValueError:
    print("\n❌ ERRO: Por favor, digite apenas números inteiros válidos.")

else:
    # 2. LÓGICA: Verifica qual número é o maior (o primeiro na ordem decrescente)

    # Caso 1: O num1 é o maior
    if num1 >= num2 and num1 >= num3:
        if num2 >= num3:
            # Ordem: num1, num2, num3
            print(f"\nOrdem decrescente: {num1}, {num2}, {num3}")
        else:
            # Ordem: num1, num3, num2
            print(f"\nOrdem decrescente: {num1}, {num3}, {num2}")

    # Caso 2: O num2 é o maior
    elif num2 >= num1 and num2 >= num3:
        if num1 >= num3:
            # Ordem: num2, num1, num3
            print(f"\nOrdem decrescente: {num2}, {num1}, {num3}")
        else:
            # Ordem: num2, num3, num1
            print(f"\nOrdem decrescente: {num2}, {num3}, {num1}")

    # Caso 3: O num3 é o maior (ou o que sobrou)
    else: # num3 é o maior
        if num1 >= num2:
            # Ordem: num3, num1, num2
            print(f"\nOrdem decrescente: {num3}, {num1}, {num2}")
        else:
            # Ordem: num3, num2, num1
            print(f"\nOrdem decrescente: {num3}, {num2}, {num1}")
```

12) Um estabelecimento está vendendo combustíveis com descontos variados. Para o etanol, se a quantidade comprada for até 15 litros, o desconto será de 2% por litro. Caso contrário, será de 4% por litro. Para o diesel, se a quantidade comprada for até 15 litros, o desconto será de 3% por litro. Caso contrário, será de 5% por litro. O preço do litro de diesel é R$ 2.00  e o preço do litro de etanol é R$ 1.70. Escreva um programa que leia a quantidade de litros vendidos e o tipo de combustível (E para etanol e D para diesel) e calcule o valor a ser pago pelo cliente. Tenha em mente algumas dicas:

O do valor do desconto será a multiplicação entre preço do litro, quantidade de litros e o valor do desconto.
O valor a ser pago por um cliente será o resultado da multiplicação do preço do litro pela quantidade de litros menos o valor de desconto resultante do cálculo.

```python
# 1. Definindo os preços fixos
PRECO_ETANOL = 1.70
PRECO_DIESEL = 2.00

# 2. Leitura da quantidade e do tipo de combustível
try:
    litros = float(input("Digite a quantidade de litros vendidos: "))
    combustivel = input("Digite o tipo de combustível (E para Etanol ou D para Diesel): ").upper()

except ValueError:
    print("\n❌ ERRO: Por favor, digite um valor numérico válido para os litros.")
    exit()

# 3. Inicializando variáveis de cálculo
preco_total_sem_desconto = 0.0
desconto_por_litro = 0.0
valor_desconto = 0.0
valor_a_pagar = 0.0

# 4. Estrutura de Decisão (if/elif/else)
if combustivel == 'E':
    # A) Lógica para ETANOL
    preco_total_sem_desconto = litros * PRECO_ETANOL

    if litros <= 15:
        # Desconto de 2% (0.02)
        desconto_por_litro = 0.02
    else:
        # Desconto de 4% (0.04)
        desconto_por_litro = 0.04

    # Cálculo do desconto e valor final (válido para Etanol)
    valor_desconto = preco_total_sem_desconto * desconto_por_litro
    valor_a_pagar = preco_total_sem_desconto - valor_desconto

elif combustivel == 'D':
    # B) Lógica para DIESEL
    preco_total_sem_desconto = litros * PRECO_DIESEL

    if litros <= 15:
        # Desconto de 3% (0.03)
        desconto_por_litro = 0.03
    else:
        # Desconto de 5% (0.05)
        desconto_por_litro = 0.05

    # Cálculo do desconto e valor final (válido para Diesel)
    valor_desconto = preco_total_sem_desconto * desconto_por_litro
    valor_a_pagar = preco_total_sem_desconto - valor_desconto

else:
    # C) Tratamento de entrada inválida
    print("\n⚠️ Tipo de combustível inválido. Por favor, digite 'E' ou 'D'.")
    exit() # Encerra o programa se o combustível for inválido

# 5. Imprime o resultado final (executado apenas se o combustível for 'E' ou 'D')
if combustivel in ('E', 'D'):
    print("\n--- RESUMO DA COMPRA ---")
    print(f"Tipo de Combustível: {combustivel}")
    print(f"Quantidade de Litros: {litros:.2f} L")
    print(f"Preço Bruto: R$ {preco_total_sem_desconto:.2f}")
    print(f"Desconto Aplicado: {desconto_por_litro * 100:.0f}%")
    print(f"Valor do Desconto: R$ {valor_desconto:.2f}")
    print(f"VALOR TOTAL A PAGAR: R$ {valor_a_pagar:.2f}")
```

13) Em uma empresa de venda de imóveis você precisa criar um código que analise os dados de vendas anuais para ajudar a diretoria na tomada de decisão. O código precisa coletar os dados de quantidade de venda durante os anos de 2022 e 2023 e fazer um cálculo de variação percentual. A partir do valor da variação, deve ser enviada às seguintes sugestões:

Para variação acima de 20%: bonificação para o time de vendas.
Para variação entre 2% e 20%: pequena bonificação para time de vendas.
Para variação entre 2% e -10%: planejamento de políticas de incentivo às vendas.
Para bonificações abaixo de -10%: corte de gastos.

```python

venda_2022 = float(input('Informe a quantidade de vendas em 2022: '))
venda_2023 = float(input('Informe a quantidade de vendas em 2023: '))

# Calculamos a variação percentual entre as vendas dos anos de 2022 e 2023
var_percentual = 100 * (venda_2023 - venda_2022) / (venda_2022)

# Análise condicional da variação percentual para determinar a sugestão a ser enviada
if var_percentual > 20:
    print('Bonificação para o time de vendas.')
elif 2 <= var_percentual <= 20:
    print('Pequena bonificação para time de vendas.')
elif -10 <= var_percentual < 2:
    print('Planejamento de políticas de incentivo às vendas.')
else:
    print('Corte de gastos.')
```

2) Escreva um programa para calcular quantos dias levará para a colônia de uma bactéria A ultrapassar ou igualar a colônia de uma bactéria B, com base nas taxas de crescimento de 3% e 1,5% respectivamente. Considere que a colônia A inicia com 4 elementos e a B com 10.

```python
colonia_a = 4
colonia_b = 10

taxa_a = 0.03
dias = 0 # Inicialização do contador de dias
taxa_b = 0.015

while colonia_a <= colonia_b:
  dias += 1
  colonia_a += colonia_a * taxa_a
  colonia_b += colonia_b * taxa_b

print(f"A colônia A ultrapassará (ou igualará) a colônia B em **{dias} dias**.")
print(f"Neste dia ({dias}):")

print(f"  Colônia A terá aproximadamente: **{colonia_a:.2f}** elementos")
print(f"  Colônia B terá aproximadamente: **{colonia_b:.2f}** elementos")
```

Encerre a entrada de dados com um número negativo.

```python
print("--- Distribuição de Idades de Pensionistas ---")
print("Digite um número negativo para encerrar a entrada de dados.")

# Inicializa os contadores para cada faixa etária
c_0_25 = 0
c_26_50 = 0
c_51_75 = 0
c_76_100 = 0
total_clientes = 0

# Inicializa a variável 'idade' com um valor não negativo para entrar no loop
idade = 0

# O loop continua enquanto a idade digitada não for negativa
while idade >= 0:

    # 1. Entrada de Dados
    # Assume-se que a entrada será um número (float ou int).
    idade = float(input('Digite a idade do cliente: '))

    # 2. Verifica o critério de parada
    if idade < 0:
        # Se for negativo, encerra o loop (o while checará a condição na próxima volta)
        break

    # 3. Processamento/Contagem
    # Garante que apenas idades até 100 anos serão processadas para as faixas
    if idade <= 100:

        # O uso de 'elif' (else if) garante que apenas uma condição será verdadeira por vez
        if 0 <= idade <= 25:
            c_0_25 += 1
        elif 26 <= idade <= 50:
            c_26_50 += 1
        elif 51 <= idade <= 75:
            c_51_75 += 1
        # Se a idade estiver entre 76 e 100
        elif 76 <= idade <= 100:
            c_76_100 += 1

        # Contagem total de clientes válidos (não negativos e até 100)
        total_clientes += 1

    else:
        # Trata idades maiores que 100, se for o caso
        print(f"Idade {idade:.0f} ignorada. Consideramos apenas clientes até 100 anos para as faixas.")


# --- Resultado Final ---
print("\n--- Distribuição de Clientes ---")

if total_clientes > 0:
    print(f"Total de clientes analisados (0-100 anos): {total_clientes}")
    print("-" * 35)
    print(f"Faixa [0-25]:    {c_0_25} clientes")
    print(f"Faixa [26-50]:   {c_26_50} clientes")
    print(f"Faixa [51-75]:   {c_51_75} clientes")
    print(f"Faixa [76-100]:  {c_76_100} clientes")
else:
    print("Nenhuma idade válida foi inserida.")
```

9) Em uma eleição para gerência em uma empresa com 20 pessoas colaboradoras, existem quatro candidatos(as). Escreva um programa que calcule o(a) vencedor(a) da eleição. A votação ocorreu da seguinte maneira:

Cada colaborador(a) votou em uma das quatro pessoas candidatas (que representamos pelos números 1, 2, 3 e 4).
Também foram contabilizados os votos nulos (representados pelo número 5) e os votos em branco (representados pelo número 6).
Ao final da votação, o programa deve exibir o número total de votos para cada candidato(a), os nulos e os votos em branco. Além disso, deve calcular e exibir a porcentagem de votos nulos em relação ao total de votos e a porcentagem de votos em branco em relação ao total de votos.

```python
# Contadores para cada candidato(a), voto nulo e voto em branco
votos_candidato_1 = 0
votos_candidato_2 = 0
votos_candidato_3 = 0
votos_candidato_4 = 0
votos_nulos = 0
votos_brancos = 0

TOTAL_ELEITORES = 20

print("--- Simulação de Eleição para Gerência ---")
print("Votação: [1, 2, 3, 4] = Candidatos | [5] = Nulo | [6] = Branco")

# Laço FOR para garantir exatamente 20 votos
for i in range(1, TOTAL_ELEITORES + 1):

    # Validação da entrada (sem try-except, assumimos que será um número)
    voto = int(input(f'Vote do colaborador(a) #{i} (1 a 6): '))

    # Estrutura condicional para contagem
    if voto == 1:
        votos_candidato_1 += 1
    elif voto == 2:
        votos_candidato_2 += 1
    elif voto == 3:
        votos_candidato_3 += 1
    elif voto == 4:
        votos_candidato_4 += 1
    elif voto == 5:
        votos_nulos += 1
    elif voto == 6:
        votos_brancos += 1
    else:
        # Trata votos inválidos (qualquer número fora do 1-6)
        print("Voto inválido! O voto foi anulado.")
        votos_nulos += 1 # Contabilizamos como nulo por regra

# --- Apuração e Resultados ---

# 1. Total de votos deve ser igual ao total de eleitores
total_votos_contados = TOTAL_ELEITORES

print("\n" + "=" * 30)
print("       RESULTADO DA APURAÇÃO")
print("=" * 30)

# 2. Exibe o número total de votos para cada candidato(a) e nulos/brancos
print("\n--- Total de Votos por Opção ---")
print(f"Candidato 1: {votos_candidato_1} votos")
print(f"Candidato 2: {votos_candidato_2} votos")
print(f"Candidato 3: {votos_candidato_3} votos")
print(f"Candidato 4: {votos_candidato_4} votos")
print(f"Nulos (5):   {votos_nulos} votos")
print(f"Brancos (6): {votos_brancos} votos")
print("-" * 30)
print(f"TOTAL GERAL: {total_votos_contados} votos")


# 3. Cálculo das Porcentagens
if total_votos_contados > 0:
    # Fórmula: (Votos da Categoria / Total de Votos) * 100

    porc_nulos = (votos_nulos / total_votos_contados) * 100
    porc_brancos = (votos_brancos / total_votos_contados) * 100

    print("\n--- Porcentagens ---")
    print(f"Percentual de Votos Nulos:  **{porc_nulos:.2f}%**")
    print(f"Percentual de Votos Brancos: **{porc_brancos:.2f}%**")


    # 4. Determinação do Vencedor(a)

    # Cria um dicionário para facilitar a comparação
    votos_candidatos = {
        1: votos_candidato_1,
        2: votos_candidato_2,
        3: votos_candidato_3,
        4: votos_candidato_4
    }

    # Encontra o candidato(a) com o maior número de votos
    vencedor = max(votos_candidatos, key=votos_candidatos.get)
    votos_vencedor = votos_candidatos[vencedor]

    # Verifica se há empate no topo
    empate = [c for c, v in votos_candidatos.items() if v == votos_vencedor]

    print("\n--- Vencedor(a) ---")
    if len(empate) > 1:
        candidatos_empatados = ", ".join(map(str, empate))
        print(f"Houve um empate! Os candidatos empatados são: {candidatos_empatados}, com {votos_vencedor} votos.")
    else:
        print(f"O(a) vencedor(a) da eleição é o(a) **Candidato(a) {vencedor}** com {votos_vencedor} votos!")

else:
    print("Erro: Não houve votos para serem contabilizados.")
```

5) Faça um programa que, ao inserir um número qualquer, cria uma lista contendo todos os números primos entre 1 e o número digitado.

```python
def eh_primo(n):
    """
    Função auxiliar para determinar se um número é primo.
    Retorna True se for primo, False caso contrário.
    """
    if n <= 1:
        return False
    # Itera de 2 até n-1. Se encontrar um divisor, não é primo.
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print("--- Gerador de Lista de Números Primos ---")
limite = int(input('Digite um número inteiro positivo (limite da busca): '))
lista_primos = []

if limite > 1:
    # 3. Itera de 2 até o número limite (inclusive)
    for num in range(2, limite + 1):
        if eh_primo(num):
            lista_primos.append(num)

    # 4. Imprime o resultado
    print(f"\n--- Números Primos entre 1 e {limite} ---")
    print(f"Total de {len(lista_primos)} números primos encontrados.")
    print(lista_primos)

elif limite == 1:
    print("\nO número 1 não possui primos menores ou iguais a ele (primos começam em 2).")

# Caso o usuário digite 0 ou um número negativo
else:
    print(f"\nO número {limite} é muito baixo. Não há números primos em um intervalo de 1 a {limite}.")
```

6) Escreva um programa que peça uma data informando o dia, mês e ano e determine se ela é válida para uma análise.

```python
def eh_bissexto(ano):
    """Verifica se o ano é bissexto."""
    # Regra: Divisível por 4, exceto por 100, a menos que seja divisível por 400.
    if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
        return True
    return False

def data_valida(dia, mes, ano):
    """Determina se a combinação de dia, mês e ano é válida."""

    # 1. Validação do Mês e Ano (Regras Básicas)
    if not (1 <= mes <= 12):
        print("Erro: O mês deve estar entre 1 e 12.")
        return False

    if ano < 1:
        # Geralmente, datas para análise não são anteriores ao ano 1
        print("Erro: O ano deve ser maior que zero (ou um valor razoável para a análise).")
        return False

    # 2. Validação dos Meses de 31 dias
    if mes in (1, 3, 5, 7, 8, 10, 12):
        if not (1 <= dia <= 31):
            print(f"Erro: O mês {mes} tem 31 dias. O dia deve estar entre 1 e 31.")
            return False

    # 3. Validação dos Meses de 30 dias
    elif mes in (4, 6, 9, 11):
        if not (1 <= dia <= 30):
            print(f"Erro: O mês {mes} tem 30 dias. O dia deve estar entre 1 e 30.")
            return False

    # 4. Validação de Fevereiro (Mês 2)
    elif mes == 2:
        if eh_bissexto(ano):
            # Ano bissexto, Fevereiro tem 29 dias
            if not (1 <= dia <= 29):
                print(f"Erro: O ano {ano} é bissexto. Fevereiro tem 29 dias. Dia inválido.")
                return False
        else:
            # Ano não bissexto, Fevereiro tem 28 dias
            if not (1 <= dia <= 28):
                print(f"Erro: O ano {ano} não é bissexto. Fevereiro tem 28 dias. Dia inválido.")
                return False

    return True


print("--- Validador de Data para Análise ---")

# 1. Coleta das Entradas
# ATENÇÃO: Se o usuário digitar texto, o programa falhará aqui (ValueError).
dia = int(input('Digite o DIA (dd): '))
mes = int(input('Digite o MÊS (mm): '))
ano = int(input('Digite o ANO (aaaa): '))

# 2. Chamada da Validação
if data_valida(dia, mes, ano):
    print(f"\nA data {dia:02d}/{mes:02d}/{ano} é **VÁLIDA** para análise.")
else:
    print(f"\nA data {dia:02d}/{mes:02d}/{ano} é **INVÁLIDA** para análise. Verifique as regras.")
```

10) Um instituto de meteorologia deseja fazer um estudo de temperatura média de cada mês do ano. Para isso, você precisa fazer um código que colete e armazene essas temperaturas médias em uma lista. Depois, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual e em que mês elas ocorreram, mostrando os meses por extenso (Janeiro, Fevereiro, etc.).

```python

meses_por_extenso = [
    "Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho",
    "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"
]

temperaturas_coletadas = []
NUMERO_DE_MESES = 12
soma_temperaturas = 0.0

print("Por favor, digite a temperatura média para cada um dos 12 meses.")

for i in range(NUMERO_DE_MESES):
    mes_atual = meses_por_extenso[i]
    temp = float(input(f'Temperatura média de {mes_atual}: '))
    temperaturas_coletadas.append(temp)
    soma_temperaturas += temp


# --- 2. Cálculo da Média Anual ---
# Usamos a soma acumulada e o número total de meses (12)
media_anual = soma_temperaturas / NUMERO_DE_MESES

print("\n" + "=" * 50)
print(f"MÉDIA ANUAL DE TEMPERATURA: **{media_anual:.2f}°C**")
print("=" * 50)

# --- 3. Exibição das Temperaturas Acima da Média ---
print("\nMeses com Temperatura Acima da Média Anual:")

# Itera sobre a lista de temperaturas. O 'i' será o índice (0 a 11)
for i in range(NUMERO_DE_MESES):

    temperatura_mes = temperaturas_coletadas[i]

    # Compara a temperatura do mês com a média anual
    if temperatura_mes > media_anual:

        # Recupera o nome do mês usando o índice 'i'
        nome_mes = meses_por_extenso[i]

        print(f"• {nome_mes}: {temperatura_mes:.2f}°C")

print("\nAnálise concluída.")
```

11) Uma empresa de e-commerce está interessada em analisar as vendas dos seus produtos. Os dados das vendas foram armazenados em um dicionário:

{'Produto A': 300, 'Produto B': 80, 'Produto C': 60,
 'Produto D': 200, 'Produto E': 250, 'Produto F': 30}

Escreva um código que calcule o total de vendas e o produto mais vendido.

12) Uma pesquisa de mercado foi feita para decidir qual design de marca infantil mais agrada as crianças. A pesquisa foi feita e o votos computados podem ser observados abaixo:

'''
Tabela de votos da marca
Design 1 - 1334 votos
Design 2 - 982 votos
Design 3 - 1751 votos
Design 4 - 210 votos
Design 5 - 1811 votos
'''
Copiar código
Adapte os dados fornecidos para uma estrutura de dicionário. A partir dele, informe o design vencedor e a porcentagem de votos recebidos.

```python

votos_designs = {
    'Design 1': 1334,
    'Design 2': 982,
    'Design 3': 1751,
    'Design 4': 210,
    'Design 5': 1811
}

print("--- Análise de Pesquisa de Mercado (Design de Marca) ---")
total_votos = sum(votos_designs.values())
design_vencedor = max(votos_designs, key=votos_designs.get)
votos_vencedor = votos_designs[design_vencedor]
if total_votos > 0:
    # Fórmula: (Votos Vencedor / Total de Votos) * 100
    porcentagem_vencedor = (votos_vencedor / total_votos) * 100
else:
    porcentagem_vencedor = 0 # Previne divisão por zero

# --- 5. Exibir Resultados ---
print("\n" + "=" * 45)
print("             RESULTADO DA PESQUISA")
print("=" * 45)
print(f"Total de Votos Computados: **{total_votos}**")
print("-" * 45)
print(f"O Design VENCEDOR é: **{design_vencedor}**")
print(f"Votos Recebidos: {votos_vencedor}")
print(f"Porcentagem do Total: **{porcentagem_vencedor:.2f}%**")
print("=" * 45)
```

13) As pessoas colaboradoras de um setor da empresa que você trabalha vão receber um abono correspondente a 10% do salário devido ao ótimo desempenho do time. O setor financeiro solicitou sua ajuda para a verificação das consequências financeiras que esse abono irá gerar nos recursos. Assim, foi encaminhada para você uma lista com os salários que receberão o abono: [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]. O abono de cada colaborador(a) não pode ser inferior a 200. Em código, transforme cada um dos salários em chaves de um dicionário e o abono de cada salário no elemento. Depois, informe o total de gastos com o abono, quantos(as) colaboradores(as) receberam o abono mínimo e qual o maior valor de abono fornecido.

```python
salarios = [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]
abono_minimo = 200
taxa_abono = 0.10  # 10%

# Dicionário para armazenar {salário: abono}
abonos_por_salario = {}

# Contadores e acumuladores
total_gastos_abono = 0.0
colaboradores_minimo = 0
maior_abono = 0.0

print("--- Análise Financeira de Abono Salarial ---")
print(f"Total de colaboradores(as): {len(salarios)}")
print("-" * 50)

# 1. Processamento dos Salários
for salario in salarios:
    abono_calculado = salario * taxa_abono
    if abono_calculado < abono_minimo:
        abono_final = abono_minimo
        colaboradores_minimo += 1
    else:
        abono_final = abono_calculado

    # 4. Criação do Dicionário
    # Salário como chave (int) e abono final como elemento (float)
    abonos_por_salario[salario] = abono_final

    # 5. Acumulação e Determinação do Máximo
    total_gastos_abono += abono_final

    if abono_final > maior_abono:
        maior_abono = abono_final

    # Opcional: Mostrar o abono de cada salário
    # print(f"Salário: R${salario:7.2f} -> Abono: R${abono_final:7.2f}")


# 6. Exibição dos Resultados Finais
print("\n" + "=" * 50)
print("             RELATÓRIO FINANCEIRO")
print("=" * 50)
print("Dicionário {Salário: Abono}:")
print(abonos_por_salario)
print("-" * 50)
print(f"Total de Gastos com Abono:        **R${total_gastos_abono:10.2f}**")
print(f"Colaboradores com Abono Mínimo:   **{colaboradores_minimo:10}**")
print(f"Maior Valor de Abono Fornecido:   **R${maior_abono:10.2f}**")
print("=" * 50)
```

14) Uma equipe de cientistas de dados está estudando a diversidade biológica em uma floresta. A equipe fez a coleta de informações sobre o número de espécies de plantas e animais em cada área dessa floresta e armazenou essas informações em um dicionário. Nele, a chave descreve a área dos dados e os valores nas listas correspondem às espécies de plantas e animais nas áreas, respectivamente.

{'Área Norte': [2819, 7236],
 'Área Leste': [1440, 9492],
 'Área Sul': [5969, 7496],
 'Área Oeste': [14446, 49688],
 'Área Centro': [22558, 45148]}
Copiar código
Escreva um código para calcular a média de espécies por área e identificar a área com a maior diversidade biológica. Dica: use as funções built-in sum() e len().

```python
dados_diversidade = {
 'Área Norte': [2819, 7236],
 'Área Leste': [1440, 9492],
 'Área Sul': [5969, 7496],
 'Área Oeste': [14446, 49688],
 'Área Centro': [22558, 45148]
}

total_especies_geral = 0
total_areas = len(dados_diversidade)

maior_diversidade_valor = -1 # Começa com um valor impossível
area_maior_diversidade = ""

print("--- Análise de Diversidade Biológica por Área ---")
print("-" * 50)

# 1. Processamento de Cada Área
for area, especies in dados_diversidade.items():

    # 'especies' é a lista, ex: [2819, 7236]

    # a) Calcula o Total de Espécies na Área (soma dos valores da lista)
    total_especies_area = sum(especies)

    # b) Acumula o total geral para a média final
    total_especies_geral += total_especies_area

    # c) Identifica a Maior Diversidade
    if total_especies_area > maior_diversidade_valor:
        maior_diversidade_valor = total_especies_area
        area_maior_diversidade = area

    # Opcional: Imprime o total por área
    print(f"Total de espécies na {area}: {total_especies_area}")


# 2. Cálculo da Média de Espécies por Área
# Média = Soma total de todas as espécies / Número total de áreas
media_especies_por_area = total_especies_geral / total_areas


# 3. Exibição dos Resultados
print("\n" + "=" * 50)
print("             RELATÓRIO DE DIVERSIDADE")
print("=" * 50)
print(f"Média Geral de Espécies por Área: **{media_especies_por_area:.2f}**")
print("-" * 50)
print(f"Área com a Maior Diversidade Biológica:")
print(f"  - Área: **{area_maior_diversidade}**")
print(f"  - Total de Espécies: **{maior_diversidade_valor}**")
print("=" * 50)
```

15) O setor de RH da sua empresa te pediu uma ajuda para analisar as idades de colaboradores(as) de 4 setores da empresa. Para isso, foram fornecidos os seguintes dados:

{'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}
Copiar código
Sabendo que cada setor tem 10 colaboradores(as), construa um código que calcule a média de idade de cada setor, a idade média geral entre todos os setores e quantas pessoas estão acima da idade média geral.

```python
dados_setores = {
 'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
}

NUMERO_COLABORADORES_POR_SETOR = 10
NUMERO_DE_SETORES = len(dados_setores)

# Variáveis para o cálculo da média geral
soma_idades_geral = 0
total_colaboradores = 0

# Dicionário para armazenar as médias por setor
medias_por_setor = {}

print("--- Análise de Idades dos Colaboradores por Setor ---")
print("-" * 50)

# 1. Calcular a Média de Idade de Cada Setor e Acumular Soma Geral
for setor, idades in dados_setores.items():

    # Cálculo da Média Setorial
    soma_setor = sum(idades)
    media_setorial = soma_setor / NUMERO_COLABORADORES_POR_SETOR

    # Armazena a média no novo dicionário
    medias_por_setor[setor] = media_setorial

    # Acumula para o cálculo da média geral
    soma_idades_geral += soma_setor
    total_colaboradores += len(idades)

    print(f"Média de idade do {setor}: {media_setorial:.2f} anos")

# 2. Calcular a Idade Média Geral
idade_media_geral = soma_idades_geral / total_colaboradores

print("\n" + "=" * 50)
print(f"IDADE MÉDIA GERAL DA EMPRESA: **{idade_media_geral:.2f} anos**")
print("=" * 50)


# 3. Contar Quantas Pessoas Estão Acima da Média Geral
acima_da_media_geral = 0

# Precisamos iterar sobre CADA IDADE em CADA SETOR
for idades in dados_setores.values():
    for idade in idades:
        if idade > idade_media_geral:
            acima_da_media_geral += 1

# 4. Exibir o Resultado Final
print(f"\nNúmero de Colaboradores Acima da Média Geral ({idade_media_geral:.2f} anos):")
print(f"**{acima_da_media_geral} pessoas**")
print("-" * 50)
print(f"Média de Idade por Setor: {medias_por_setor}")
```
