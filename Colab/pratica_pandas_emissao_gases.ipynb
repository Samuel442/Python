Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a soma de emissão de cada setor. Ordene os dados da maior para menor emissão.

```python
# Seleciona todas as colunas de anos (que são numéricas)
colunas_anos = [col for col in emissoes_gases.columns if isinstance(col, int)]

# Agrupa por setor e soma todas as colunas numéricas
agrupado_soma = emissoes_gases.groupby("Nível 1 - Setor")[colunas_anos].sum()

# Cria uma coluna com o total geral de emissões por setor
agrupado_soma["Total Emissões"] = agrupado_soma.sum(axis=1)

# Ordena do maior para o menor
resultado_final = agrupado_soma.sort_values("Total Emissões", ascending=False)

resultado_final
```

Uma das tarefas passadas pela equipe de supervisão do órgão ambiental era de encontrar a emissão média de cada gás por ano. Conseguimos realizar essa tarefa usando o método groupby() e pivot_table() para gerar uma tabela contendo a emissão ao longo dos anos para cada um dos gases.

A equipe de supervisão deseja saber agora a média de emissão para cada um dos setores econômicos da coluna "Nível 1 - Setor" ao longo dos anos. Como desafio, crie uma tabela de pivô e uma visualização gráfica contendo essas informações, utilizando apenas o método pivot_table para a construção da tabela, sem utilizar o método groupby.

```python
colunas_anos = emissoes_gases.columns[emissoes_gases.columns.astype(str).str.isnumeric()]
```

```python
tabela_setores = emissoes_gases.pivot_table(
    index="Nível 1 - Setor",
    values=colunas_anos,
    aggfunc="mean"
)
```

```python
tabela_setores
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(tabela_setores.T)

plt.title("Média de Emissões ao Longo dos Anos por Setor (Nível 1)")
plt.xlabel("Ano")
plt.ylabel("Emissão Média")
plt.legend(tabela_setores.index, title="Setores")
plt.grid(True)
plt.show()
```

1) Faça um agrupamento de dados com as colunas "Estado" e "Nível 1 - Setor", obtendo a soma de emissão e armazenando o resultado em uma tabela.

```python
# selecionar apenas as colunas de anos (numéricas)
colunas_anos = [col for col in emissoes_gases.columns if isinstance(col, int)]

# agrupamento por Estado e Nível 1 - Setor
tabela_agrupada = (
    emissoes_gases.groupby(["Estado", "Nível 1 - Setor"])[colunas_anos]
    .sum()
    .reset_index()
)

tabela_agrupada
```

2) Utilizando a tabela construída na atividade 1, selecione os dados referentes à "Energia" do índice "Nível 1 - Setor".

```python
tabela_agrupada = tabela_agrupada.reset_index()
```

```python
energia = tabela_agrupada[tabela_agrupada["Nível 1 - Setor"] == "Energia"]
energia
```

3) Utilizando a tabela construída na atividade 1, encontre a atividade econômica com valor máximo de emissão do Estado de Minas Gerais

```python
# --- Questão 3: Encontrar o setor com maior emissão em MG ---

# 1) Garantir que o índice tenha "Estado" em nível superior
# Se o índice NÃO for MultiIndex, criamos um corretamente
if not isinstance(tabela_agrupada.index, pd.MultiIndex):
    tabela_agrupada = tabela_agrupada.set_index(["Estado", "Nível 1 - Setor"])

# 2) Seleciona apenas MG
mg = tabela_agrupada.loc["MG"]

# 3) Soma total de emissões por setor (somando todos os anos numéricos)
soma_setores_mg = mg.sum(axis=1)

# 4) Descobre qual setor teve a maior emissão
setor_max_mg = soma_setores_mg.idxmax()
valor_max_mg = soma_setores_mg.max()

setor_max_mg, valor_max_mg
```

4) Obtenha uma tabela contendo a atividade econômica com máxima emissão para cada Estado.

```python
# 1. Soma total de emissões de cada setor dentro de cada estado
soma_total = tabela_agrupada.sum(axis=1)   # soma todos os anos

# 2. Transformar isso em DataFrame
soma_df = soma_total.reset_index()
soma_df.columns = ["Estado", "Nível 1 - Setor", "Emissão Total"]

# 3. Para cada estado, pegar o setor com maior emissão
resultado = soma_df.loc[soma_df.groupby("Estado")["Emissão Total"].idxmax()]

resultado
```
